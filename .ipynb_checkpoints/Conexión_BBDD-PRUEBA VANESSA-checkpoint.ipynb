## Conexión y subida de los datos a la BBDD
A continuación se va a realizar la conexión con la BBDD en PostgreSQL y luego se van a subir los datos en la misma.

Conexión a la BBDD

Para instalar pymssql se coloca en el terminal del ordenador conda install pymssql


Seguir estas instrucciones https://docs.microsoft.com/es-es/sql/connect/python/pymssql/step-1-configure-development-environment-for-pymssql-python-development?view=sql-server-2017

```python
import pymssql 
```

```python
conn = pymssql.connect(server='serveraccidentes.database.windows.net', user='byludo@serveraccidentes', password='Mariop96', database='accidentes')
```

Conversión del csv en DataFrame

```python
import pandas as pd
```

```python
tipologia=pd.read_csv('Tipología Accidentes/Tipologia_Accidente.csv', index_col=False)
tipologia.head
```

Inserción de los datos en la BBDD

```python
cur = conn.cursor()
```

```python
cur.fast_executemany = True
cur.executemany("INSERT INTO tipologia_accidente(num_expediente, tipologia_accidente) VALUES (?, ?)", list(tipologia.itertuples(index=False, name=None)))
cur.commit()
```

Revisión que los datos se hayan insertado correctamente

```python
cur.execute('SELECT * FROM tipologia_accidente')
rows = cur.fetchall()
print("Número de lineas: ", cur.rowcount)
for row in rows:
     print(row)
cur.close()
```
